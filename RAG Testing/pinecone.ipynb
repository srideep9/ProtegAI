```python
from llama_index.llms.huggingface import HuggingFaceInferenceAPI
llm = HuggingFaceInferenceAPI(model_name = 'meta-llama/Llama-3.2-3B-Instruct', token = HF_TOKEN)
print(llm)
```

```python
from llama_index.core import SimpleDirectoryReader
documents = SimpleDirectoryReader(input_files=["test.txt"]).load_data()
documents[0].get_content()
```

```python
from llama_index.embeddings.huggingface import HuggingFaceEmbedding
embed_model = HuggingFaceEmbedding()
```

```python
%pip install pinecone

%pip install pinecone[grpc]

%pip install llama-index-vector-stores-pinecone
```

```python
from pinecone.grpc import PineconeGRPC
from pinecone import ServerlessSpec
from llama_index.vector_stores.pinecone import PineconeVectorStore
```

```python
pc = PineconeGRPC(api_key=PINECONE_API_KEY)
index_name = "llama-integration-example"
```

```python
#dimensions are based on huggingface embedding
pc.create_index(index_name, dimension=384, spec = ServerlessSpec(cloud = "aws", region = "us-east-1"))
```

```python
pinecone_index = pc.Index(index_name)
```

```python
vector_store = PineconeVectorStore(pinecone_index=pinecone_index)

from llama_index.core import StorageContext

storage_context = StorageContext.from_defaults(vector_store=vector_store)
```

```python
from llama_index.core import VectorStoreIndex
index = VectorStoreIndex.from_documents(documents, storage_context=storage_context, embed_model=embed_model)
```

```python
#set query engine
query_engine = index.as_query_engine(llm=llm)
```

```python
response = query_engine.query("Summarize Lack Of Support")
print(response)
```

```python
response2 = query_engine.query("What section talks about the AMLAW 200 pro bono ranking?")
print(response2)
```
